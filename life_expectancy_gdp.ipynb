```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all_data = pd.read_csv("all_data.csv")
print(all_data.head(10))
```

```python
sns.scatterplot(x = "Year", y= "Life expectancy at birth (years)", hue= "Country", data = all_data)
plt.xlabel("Year")
plt.ylabel("Life expectancy at birth (years)")
plt.title("Life expectancy over time in the six countries")
plt.legend(bbox_to_anchor= (1.0, 1.0))
plt.show()
plt.clf()
```

```python
sns.scatterplot(x= "Year", y= "GDP", hue= "Country", data= all_data)
plt.xlabel("Year")
plt.ylabel("GDP")
plt.title("GPD over time in six countries")
plt.legend(loc= "upper left")
plt.show()
plt.clf()
```

```python
sns.scatterplot(x= "Year", y= "Life expectancy at birth (years)", 
                hue= "Country", size= all_data["GDP"], data= all_data, sizes=(300, 600), alpha= 0.5)
plt.legend(bbox_to_anchor= (1.0, 1.0))
plt.show()
plt.clf()
```

```python
sns.histplot(all_data["Life expectancy at birth (years)"])
plt.title("distribution of that life expectancy")
plt.show()
plt.clf()
```

```python
avg_lifeexp = np.mean(all_data["Life expectancy at birth (years)"])
max_lifeexp = np.max(all_data["Life expectancy at birth (years)"])
min_lifeexp = np.min(all_data["Life expectancy at birth (years)"])
print(f"The average of life expectancy: {avg_lifeexp}")
print(f"The longest life expectancy: {max_lifeexp}")
print(f"The lowest life expectancy: {min_lifeexp}")
```

```python
sns.boxplot(x= "Country", y= "Life expectancy at birth (years)", data= all_data)
plt.xticks(rotation= 25)
plt.title("life expectancy distribution by country")
plt.show()
plt.clf()
```

```python
all_data.dropna(subset= ["Life expectancy at birth (years)", "GDP", "Country"], inplace=True)
graph = sns.FacetGrid(all_data, col= "Country")
graph.map(sns.scatterplot, "Life expectancy at birth (years)", "GDP")
plt.show()
plt.clf()
```

```python
graph = sns.FacetGrid(all_data, col= "Country")
graph.map(sns.lineplot, "Year", "GDP")
plt.show()
plt.clf()
```

```python
graph = sns.FacetGrid(all_data, col= "Country")
graph.map(sns.lineplot, "Year", "Life expectancy at birth (years)")
plt.show()
```
